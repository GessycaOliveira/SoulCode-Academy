# Recebendo e guardando as sequências numéricas do usuário.

```python
print('Digite 2 listas de números com a mesma quantidade de números nos campos abaixo: ')
x = input('Digite a primeira sequência numérica: ')
y = input('Digite a segunda sequência numérica: ')

# convertendo todos os numeros em float (tratamento)
lista_x = []
trat_x = x.split()

# Tente converter, caso não consiga, solicite nova sequência
try:
    for valor in trat_x: # Percorrer a lista e converter os valores para float
        lista_x.append(float(valor))

except:
    print('Erro na conversão de dados, por favor, digite apenas números.')

lista_y = []
trat_y = y.split()
try:
    for valor in trat_y:
        lista_y.append(float(valor))

except:
    print('Erro na conversão de dados, por favor, digite apenas números.')

if len(lista_x) != len(lista_y):
    print('Erro na conversão de dados, por favor, digite a mesma quantidade de números em ambas as sequências')
```

### Média aritmética

> É a somatória dos elementos dividido pela quantidade de elementos

image.png

```python
# Média aritmética X
quant_x = len(lista_x) # Criando atalho para quantidade de numeros que contem na lista x
media_x = sum(lista_x) / quant_x  # Soma dos valores / quantidade de números na variável
print(f"A média da primeira sequência é: {media_x:.2f}")
# Média aritmética Y
quant_y = len(lista_y)
media_y = sum(lista_y) / quant_y
print(f"A média da segunda sequência é: {media_y:.2f}")
```

### Mediana

> É o valor central de um conjunto de dados
> No caso se a quantidade total do conjunto de dados for:
* Ímpar: o valor central
* Par: soma e divide os 2 valores centrais

image.png

```python
# Mediana X
lista_x.sort() # Colocar números na ordem
if len(lista_x) % 2 == 0: #Se a mediana possuir 2 valores (soma da quantidade de numeros pares)
    m1 = lista_x[len(lista_x) // 2] #segunda metade
    m2 = lista_x[len(lista_x) // 2-1] #primeira metade
    mediana_x = (m1 + m2) / 2 #soma e divisão
else:
    mediana_x = lista_x[len(lista_x) // 2 ] #se tiver apenas 1 valor ao meio
print(f'A mediana da primeira sequência é: {mediana_x}')
# Mediana Y
lista_y.sort()
if len(lista_y) % 2 == 0:
    m1 = lista_y[len(lista_y) // 2]
    m2 = lista_y[len(lista_y) // 2-1]
    mediana_y = (m1 + m2) / 2
else:
    mediana_y = lista_y[len(lista_y) // 2 ]
print(f'A mediana da segunda sequência é: {mediana_y}')
```

### Moda

* Moda de X

```python
existe_moda_x = False
moda_x = ''
cont_moda_x = 0 # Qunatas vezes o nuumero se repete
cont_atual_x = 0 # Conta o número de vezes em que valor x aparece
valor_anterior_x = '' # Inicializada como str, pois é um valor que não existe na lista
for valor_x in lista_x:
    if valor_x != valor_anterior_x: # Verificar se o valor atual é diferente do anterior
        valor_anterior_x = valor_x # Caso seja, atuliza a variavel valor anterior e o contador
        cont_atual_x = 0

    cont_atual_x += 1 # Soma +1 para cada vez que um número aparece

    if cont_atual_x == cont_moda_x: # Se existir igualdade entre valores de moda, não existirá moda
        existe_moda = False

    if cont_atual_x > cont_moda_x:
        moda_x = valor_x
        cont_moda_x = cont_atual_x
        existe_moda_x = True

if existe_moda_x == False:
    print('A primeira sequência não existe moda')
else:
    print(f'A moda da primeira sequência numérica é: {moda_x}')
```

### Quartis

**O que é o boxplot?**

O boxplot ou diagrama de caixas é um método utilizado para a análise exploratória de variáveis quantitativas ou ordinais. Ele é uma ferramenta capaz de fornecer diversas medidas estatísticas como mínimo, primeiro quartil, segundo quartil (mediana), terceiro quartil, o máximo e os possíveis outliers (valores discrepantes).

image.png

* Primeira varável

```python
# divivindo lista de valores originais em duas metades
i = quant_x//2
x_inicio_meio = lista_x[0:i]
x_meio_fim = lista_x[i:]

# calculando q1
if len(x_inicio_meio) % 2 == 0:
    k = len(x_inicio_meio)//2
    v1 = x_inicio_meio[k-1] #primeira metade
    v2 = x_inicio_meio[k] #segunda metade
    q1_x = (v1 + v2) / 2 #soma e divisão
else:
    k = len(x_inicio_meio) // 2 #se tiver apenas 1 valor ao meio
    q1_x = x_inicio_meio[k]

# calculando q3
if len(x_meio_fim) % 2 == 0:
    k = len(x_meio_fim)//2
    v1 = x_meio_fim[k-1] #primeira metade
    v2 = x_meio_fim[k] #segunda metade
    q3_x = (v1 + v2) / 2 #soma e divisão
else:
    k = len(x_meio_fim) // 2 #se tiver apenas 1 valor ao meio
    q3_x = x_meio_fim[k]


#AIQ de X
aiq_x = q3_x - q1_x
print(f'''Os valores de quartis da primeira sequência:
1º quartil: {q1_x}
3º quartil: {q3_x}
Amplitude interquartílica: {aiq_x} ''')
```

* Segunda varável

```python
# divivindo lista de valores originais em duas metades
i = quant_y//2
y_inicio_meio = lista_y[0:i]
y_meio_fim = lista_y[i:]

# calculando q1
if len(y_inicio_meio) % 2 == 0:
    k = len(y_inicio_meio)//2
    v1 = y_inicio_meio[k-1]
    v2 = y_inicio_meio[k]
    q1_y = (v1 + v2) / 2
else:
    k = len(y_inicio_meio) // 2
    q1_y = y_inicio_meio[k]

# calculando q3
if len(y_meio_fim) % 2 == 0:
    k = len(y_meio_fim)//2
    v1 = y_meio_fim[k-1]
    v2 = y_meio_fim[k]
    q3_y = (v1 + v2) / 2
else:
    k = len(y_meio_fim) // 2
    q3_y = y_meio_fim[k]

#AIQ de X
aiq_y = q3_y - q1_y
print(f'''Os valores de quartis da primeira sequência:
1º quartil: {q1_y}
3º quartil: {q3_y}
Amplitude interquartílica: {aiq_y} ''')
```

### Outliers

> Os outliers indicam possíveis valores discrepantes. No boxplot, as observações são consideradas outliers quando estão abaixo ou acima do limite de detecção de outliers

```python
lista_out_x = []
c = 0
while c < len(lista_x):
    if lista_x[c] < lim_inf_x or lista_x[c] > lim_sup_x:
        lista_out_x.append(lista_x[c])
    c += 1
if len(lista_out_x) != 0:
    print(f'Outliers da primeira sequência: {lista_out_x}')
else:
    print(f'Nesta sequência não existe outliers.')

lista_out_y = []
c = 0
while c < len(lista_y):
    if lista_y[c] < lim_inf_y or lista_y[c] > lim_sup_y:
        lista_out_y.append(lista_y[c])
    c += 1
if len(lista_out_y) != 0:
    print(f'Outliers da segunda sequência: {lista_out_y}')
else:
    print(f'Nesta sequência não existe outliers.')
```

## Variância

> Mede a dispersão média dos dados em torno da média aritmética é
representada por S².

> Quanto mais longe os dados estão da média, maior o valor da variância (e vice-versa)

image.png

```python
# Variância X
var_x = []
vari_x = 0
for valor_x in lista_x: # Para cada valor na lista X execute:
    vari_x = vari_x + (valor_x - media_x) **2 # X = B + ( X - Media X)
    var_x.append(vari_x) # Adicionar na variável X
vari_x = vari_x / (quant_x - 1)
print(f'A variância da primeira sequência é: {vari_x:.2f}')
# Variância Y
var_y = []
vari_y = 0
for valor_y in lista_y:
    vari_y = vari_y + (valor_y - media_y) **2
    var_y.append(vari_y)
vari_y = vari_y / (quant_y - 1)
print(f'A variância da segunda sequência é: {vari_y:.2f}')
```

## Correlação de Pearson

> A correlação de Pearson é uma
medida padronizada que avalia a relação linear entre duas
variáveis contínuas.

> Ela mede a força e a direção da relação linear entre as variáveis, variando de -1 a 1.

image.png

```python
# Somatórias
soma_de_x = sum(lista_x)
soma_de_y = sum(lista_y)
# Elevação ao quadrado percorrendo a lista indidualmente
elev_x = 0
for i in range(len(lista_x)):
    valor_x = lista_x[i] ** 2
    elev_x = elev_x + valor_x
elev_y = 0
for i in range(len(lista_y)):
    valor_y = lista_y[i] ** 2
    elev_y = elev_y + valor_y
# Total de X.Y
total_xy = 0
for i in range(len(lista_x)):
    valor_xy = lista_x[i] * lista_y[i]
    total_xy = total_xy + valor_xy
# Equação
n = len(lista_x)
numerador_pearson = total_xy - ((soma_de_x * soma_de_y)/n)
denominador_pearson = (elev_x - ((soma_de_x)**2)/n) * (elev_y - ((soma_de_y)**2)/n)
denominador_pearson = denominador_pearson ** (1/2)
r_pearson = numerador_pearson / denominador_pearson
print(f'O coeficiênte de Pearson é: {r_pearson:.4f}')
```

> Um coeficiente de correlação de Pearson próximo de +1 indica uma correlação
positiva forte, enquanto um coeficiente próximo de -1 indica uma
correlação negativa forte.

> Um valor próximo de 0 indica uma correlação fraca ou inexistente.

## R**2 de Pearson

> Eleva o resultado da correlação de pearson ao quadrado.

image.png

```python
quad_r_pearson = r_pearson ** 2
print(f'O coeficiênte de Pearson é: {quad_r_pearson:.4f}')
```

> O valor de R² varia de 0 a 1 e é interpretado como a porcentagem da variação em y que pode ser explicada pelas variáveis independentes.

> Um valor de R² igual a 1 indica que todas as variações em y são perfeitamente
explicadas pelas variáveis independentes do modelo.

> Por outro lado, um valor de R² igual a 0 indica que as  variáveis independentes não explicam nenhuma variação na variável dependente.

## Regressão linear simples

> A regressão linear simples é um método estatístico utilizado para
modelar a relação entre duas variáveis, uma chamada variável
independente (x) e outra chamada variável dependente (y). É chamada
de "linear" porque assume que essa relação pode ser representada por
uma equação de uma linha reta.


image.png

image.png

```python
# Cálculo beta
beta = (total_xy - (soma_de_x * soma_de_y)/n) / (elev_x - (soma_de_x**2)/n)
# Cálculo de alfa
alpha = media_y - beta * media_x
print(f'O coeficiente Alpha é: {alpha:.2f}')
print(f'O coeficiente Beta é: {beta:.2f}')
print(f'A equação da regressão linear simples é: y = {alpha:.2f} + {beta:.2f} * x')
```

os valores futuros da variável dependente com base nos valores da
